## Question 1

Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.

```python
df = pd.read_csv(r"C:\Users\mila_\Documents\WLU - Y3\CP321 - Data Visualization\Assignments\energy-price.csv")
#filter to only show Regular Unleaded Gasoline within 2022 
gasoline = df[(df['Fuel Type'] == "Regular Unleaded Gasoline") & (df['Date'].str.contains('2023'))]
# list of columns to exclude
columns_to_exclude = ["_id", "Ontario Average/Moyenne provinciale", "Southern Average/Moyenne du sud de l'Ontario", "Northern Average/Moyenne du nord de l'Ontario"]
# drop the specified columns
gasoline = gasoline.drop(columns_to_exclude, axis=1)
# calculate the mean 
mean = gasoline.mean(numeric_only=True)
# sort the mean prices
mean_sorted = mean.sort_values()
# plot a bar graph 
mean_sorted.plot.bar()

plt.xlabel("Cities")
plt.ylabel("Mean Price ")
plt.title("Regular Unleaded Gasoline price among different cities in 2023")
plt.show()
```

## Question 2
Plot the corresponding dot plot for the same information presented in Question 1.

```python
# using previous filtered gasoline data and calculated mean
# plot a dot plot graph instead
plt.scatter(x=mean_sorted.index, y=mean_sorted.values, marker='o', color='blue')
# rotate x-axis values for better look
plt.xticks(rotation=90)

plt.xlabel("Cities")
plt.ylabel("Mean Price ")
plt.title("Regular Unleaded Gasoline price among different cities in 2023")
plt.show()
```

## Question 3

Plot a histogram of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023, using 15 bins.

```python
#using previous filtered gasoline data to show the price for Toronto West/Ouest
toronto = gasoline[['Date', 'Toronto West/Ouest']]
# plot histogram with 15 bins
toronto.hist(column='Toronto West/Ouest', bins=15)

plt.xlabel("Price")
plt.ylabel("Frequency ")
plt.title("Regular Unleaded Gasoline price in Toronto West/Ouest for 2023")
plt.show()
```

## Question 4

Plot the normalized ECDF (Empirical Cumulative Distribution Function) of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023.

```python
#filtering all values for Toronto West/Ouest
df = gasoline['Toronto West/Ouest'].values
#calculating ECDF
x = np.sort(df)
y = np.arange (1, len(df) + 1) / len(df)
#plotting 
plt.figure(figsize=(10, 6))
plt.step(x, y) 
plt.title("ECDF Normalized of the R.U.G. price in Toronto West/Ouest for 2023")
plt.xlabel("Price")
plt.ylabel("ECDF")
plt.show()
```

## Question 5

Plot a boxplot to display the distribution of Regular Unleaded Gasoline prices across different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") from 2021 onwards (including 2021).


```python
df = pd.read_csv(r"C:\Users\mila_\Documents\WLU - Y3\CP321 - Data Visualization\Assignments\energy-price.csv")
#filter to only show Regular Unleaded Gasoline from 2021 & onwards
gasoline = df[(df['Fuel Type'] == "Regular Unleaded Gasoline") & (df['Date'].str[:4].astype(int) >= 2021)]
# list of columns to exclude
columns_to_exclude = ["_id", "Ontario Average/Moyenne provinciale", "Southern Average/Moyenne du sud de l'Ontario", "Northern Average/Moyenne du nord de l'Ontario"]
# drop the specified columns
gasoline = gasoline.drop(columns_to_exclude, axis=1)
# plot a boxplot
gasoline.boxplot()
# rotate x-axis values for better look
plt.xticks(rotation=90)

plt.title("Regular Unleaded Gasoline price from 2021 and onward")
plt.xlabel("Cities")
plt.ylabel("Prices")
plt.show()
```
